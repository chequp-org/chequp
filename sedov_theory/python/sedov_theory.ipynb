```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
j = 2 # cylindrical
w = 0 # uniform background plasma
gamma = 5./3
```

```python
a = (j+2-w)*(gamma+1)/4
b = (gamma+1)/(gamma-1)
c = (j+2-w)*gamma/2
d = (j+2-w)*(gamma+1)/( (j+2-w)*(gamma+1) - 2*(2+j*(gamma-1)) )
e = (2+j*(gamma-1))/2
```

```python
alpha_0 = 2/(j+2-w)
alpha_2 = - (gamma-1)/( 2*(gamma-1) + j - gamma*w )
alpha_1 = (j+2-w)*gamma/(2+j*(gamma-1)) * ( 2*(j*(2-gamma)-w)/(gamma*(j+2-w)**2) - alpha_2 )
alpha_3 = (j-w)/( 2*(gamma-1) + j - gamma*w )
alpha_4 = (j+2-w)*(j-w)/( j*(2-gamma) - w ) * alpha_1
alpha_5 = ( w*(1+gamma) - 2*j )/( j*(2-gamma) - w )
```

```python
V0 = 2/((j+2-w)*gamma)
V2 = 4/((j+2-w)*(gamma+1))
```

```python
def j1_integrand( V ):
    return - (gamma+1)/(gamma-1) * V**2 * ( alpha_0/V + alpha_2*c/(c*V-1) - alpha_1*e/(1-e*V) ) \
            * ( (a*V)**alpha_0 * (b*(c*V-1))**alpha_2 * (d*(1-e*V))**alpha_1 )**( -(j+2-w) ) \
            * (b*(c*V-1))**alpha_3 * (d*(1-e*V))**alpha_4 * (b*(1-c*V/gamma))**alpha_5
```

```python
def j2_integrand( V ):
    return - (gamma+1)/(2*gamma) * V**2 * ((c*V-gamma)/(1-c*V)) * ( alpha_0/V + alpha_2*c/(c*V-1) - alpha_1*e/(1-e*V) ) \
            * ( (a*V)**alpha_0 * (b*(c*V-1))**alpha_2 * (d*(1-e*V))**alpha_1 )**( -(j+2-w) ) \
            * (b*(c*V-1))**alpha_3 * (d*(1-e*V))**alpha_4 * (b*(1-c*V/gamma))**alpha_5
```

```python
# Evaluate the J1 and J2 integral
# Note: J2 integrand diverges near V=V0, and thus requires very fine sampling
N_pts = 10**8
V = np.linspace(V0+1.e-10, V2, N_pts)

#plt.plot( V, j1_integrand(V), label='j1_integrand' )
#plt.plot( V, j2_integrand(V), label='j2_integrand' )
#plt.legend(loc=0)
#plt.ylim(0,4)


dV = (V2-V0)/N_pts
J1 = (j1_integrand(V)*dV).sum()
J2 = (j2_integrand(V)*dV).sum()
```

```python
assert j==2
I1 = np.pi*J1
I2 = 2/(gamma-1)*np.pi*J2
alpha = I1+I2
```
